```python
hsi = df[df["HOSPITALIZACION"] == 1]
hno  = df[df["HOSPITALIZACION"] == 0]
print(hsi.shape)
print(hno.shape)
```

```python
from sklearn.utils import resample
hsi_upsample = resample(hsi,
             replace=True,
             n_samples=len(hno),
             random_state=42)

print(hsi_upsample.shape)
```

```python
df_upsampled=pd.concat([hsi_upsample, hno])
```

```python
df2=df_upsampled.drop(columns=['EDAD','DIABETES','PSA','VOLUMEN PROSTATICO'])
```

```python
df2.corr(numeric_only=False)
```

```python
X=df2.drop(columns='HOSPITALIZACION')
y=df2['HOSPITALIZACION']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
```

```python
clf.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score 
```

```python
y_pred = clf.predict(X_test)
```

```python
print('El score del modelo es:', accuracy_score(y_test, y_pred))
```

```python
print(clf.coef_, clf.intercept_)
```
